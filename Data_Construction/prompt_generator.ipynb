```python
import openai
import json
import csv
import ast
```

```python
with open("prompt.txt", "r") as f:
    prompt = f.read()
prompt
```

```python
name_list = []
with open('pokemon_preprocessed.csv','r') as csvfile:
      rdr = csv.DictReader(csvfile)
      for i in rdr:
            name = i['name']
            name_list.append(name)
```

```python
name_list = error
```

```python
for name in name_list:
  count += 1
  print(f'count : {count} , error : {error}')
  try_cnt = 0
  while True:
    try:
      new_prompt = prompt.replace('[name_of_pokemon]', name)
      response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "answer must be available as input of ast.literal_eval "},
            {"role": "system", "content": "You can only answer in form of a single python list whose length is 20"},
            {"role": "user", "content": new_prompt},
        ]
      )
      print(response['choices'][0]['message']['content'])
      desc_list = ast.literal_eval(response['choices'][0]['message']['content'])
      if len(desc_list) != 20:
        try_cnt += 1
        continue
      temp = {'name' : f'{name}', 'prompts' : desc_list}
      print(temp)
      with open("pokemon_prompts.csv", "a", newline="") as csvfile:
        csv_columns = ['name', 'prompts']
        dictwriter_object = csv.DictWriter(csvfile, fieldnames=csv_columns)
        dictwriter_object.writerow(temp)
        csvfile.close()
      break
    except:
      if try_cnt > 3:
        print(name)
        error.append(name)
        break
      else : 
        try_cnt += 1
    
   
```
